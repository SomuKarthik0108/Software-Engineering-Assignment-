```python
with open("single_input.txt", "w") as file:
    file.write("25")
```

```python
try:
    with open("single_input.txt", "r") as file:
        num = int(file.read())
    print("Number =", num)
    print("Square =", num * num)
except:
    print("Error reading file")
```

```python
with open("multiple_inputs.txt", "w") as file:
    file.write("10\n20\n30\n40")
```

```python
try:
    with open("multiple_inputs.txt", "r") as file:
        numbers = file.readlines()

    numbers = [int(n.strip()) for n in numbers]

    print("Numbers:", numbers)
    print("Sum:", sum(numbers))
    print("Average:", sum(numbers)/len(numbers))
except:
    print("Error reading file")
```

```python
try:
    x = int(input("Enter a number: "))
    print(100 / x)
except ZeroDivisionError:
    print("Cannot divide by zero")
except ValueError:
    print("Invalid input")
```

```python
files = {
    "hardcoded.py": "a=10\nb=20\nprint(a+b)",
    "keyboard.py": "a=int(input())\nb=int(input())\nprint(a+b)",
    "single_file.py": "f=open('single_input.txt')\nprint(f.read())",
    "multiple_file.py": "f=open('multiple_inputs.txt')\nprint(f.readlines())"
}

for name, content in files.items():
    with open(name, "w") as f:
        f.write(content)

print("All versions saved")
```
